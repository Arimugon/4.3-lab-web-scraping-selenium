####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
url10 = 'https://www.emsc-csem.org/Earthquake/'

driver10=webdriver.Chrome(PATH, options=opciones)

driver10.get(url10) 
````

````python
tabla2=driver10.find_element(By.XPATH, '//*[@id="tbody"]')
````

````python
data=[[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in tabla2.find_elements(By.TAG_NAME, 'tr')]

data2=data[:20]
data2
````

````python
df=pd.DataFrame(data, columns=['col1','col2','col3','col4','col5','col6','col7','col8','col9','col10','col11','col12','col13'])

#df[:21]
df=df[:21]
df
````

````python
cols=['col1','col2','col3','col10','col11','col13']
df=df.drop(cols, axis='columns')
df
````

````python
df['Latitud'] = df[['col5', 'col6']].agg(' '.join, axis=1)
df['Longitud'] = df[['col7', 'col8']].agg(' '.join, axis=1)
df
````

````python
cols=['col5','col6','col7','col8','col9']
df=df.drop(cols, axis='columns')
df
````

````python
df.rename(columns={'col4':'Date&Time',
                        'col5':'Region_Name'},
               inplace=True)
df
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
url8= 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
driver8=webdriver.Chrome(PATH, options=opciones)

driver8.get(url8) 
````

````python
tabla=driver8.find_element(By.TAG_NAME, 'tbody')

filas=tabla.find_elements(By.TAG_NAME, 'tr')



data=[[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in tabla.find_elements(By.TAG_NAME, 'tr')]

data[:2]
````

````python
cabeceras=driver8.find_element(By.TAG_NAME, 'thead')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'th')]

driver.quit()

cabeceras
````

````python
df=pd.DataFrame(data, columns=cabeceras)

df
````
